```python
import numpy as np
import ttp as ttp
```

```python
model_names = {"vgg16":"s",
                "resnet18":"v",
                "resnet50":"^",
                "resnet101":"<",
                "resnet152":">",
                "mobile_v2":"P",
                "densenet121":"*",
                "densenet169":"h",
                "densenet201":"H"}
index = ["/content/vgg16_cifar10.npy",
         "/content/resnet18_cifar10.npy",
         "/content/resnet50_cifar10.npy",
         "/content/resnet101_cifar10.npy",
         "/content/resnet152_cifar10.npy",
         "/content/mobilet_cifar10.npy",
         "/content/densenet121_cifar10.npy",
         "/content/densenet169_cifar10.npy",
         "/content/densenet201_cifar10.npy",]
accuracy_model=[94.74,95.42,95.78,95.92,95.93,95.17,96.53,96.21,96.39]
layer = [11,10,16,33,49,18,11,11,11]
```

```python
Betti_numbers = [np.load(index[0])]
for i in index[1:]:
  Betti_numbers.append(np.load(i))
print(len(Betti_numbers))
# There are 10 class 200 samples and 13 layers and betti numbers upto dim 2
Betti_numbers[0].shape
```

```python
T = ttp.TTP(degree = 3,vector = [0,1,0.5,0.5])
T.get_score(Betti_numbers,layer)
T.marker_plot(accuracy_model,model_names)
```
